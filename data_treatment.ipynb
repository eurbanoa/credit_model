```python
# Import Libs to start session
from config1 import connection_parameters
from snowflake.snowpark import Session

# Libs to use pandas for snowpark
# prerequisite to run in terminal: pip install "snowflake-snowpark-python[modin]"
import modin.pandas as pd
import snowflake.snowpark.modin.plugin
```

```python
sesion = Session.builder.configs(connection_parameters).create()
if sesion != None:
    print("Conectado")
    print(sesion.sql("select current_warehouse(), current_database(), current_role()").collect()) 

else:
    print("Error de conexión")
```

```python
# Ensure you're using the correct database and schema
sesion.use_database("CREDIT_RISK")
sesion.use_schema("PUBLIC")

credit_data_pd = pd.read_snowflake("CREDIT_DATA")
```

```python
# Show the first 5 rows
credit_data_pd
```

```python
# Set the second row as header
credit_data_pd.columns = credit_data_pd.iloc[0]  # Set second row as the header
credit_data_pd = credit_data_pd.drop(0)  # Drop the original row with placeholder headers

# Reset the index to clean up the DataFrame
credit_data_pd = credit_data_pd.reset_index(drop=True)

# Display the first few rows to verify
credit_data_pd.head()
```
